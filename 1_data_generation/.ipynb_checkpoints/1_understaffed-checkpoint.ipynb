```python
from prophet import Prophet
```

```python
# Get unique facilities
facilities = understaffed_df['facility_id'].unique()
print(f"Total facilities: {len(facilities)}")

# Prepare data for first facility as test
first_facility = facilities[0]
facility_data = understaffed_df[understaffed_df['facility_id'] == first_facility].copy()

print(f"\nTesting with facility: {first_facility}")
print(f"Records: {len(facility_data)}")
print(f"Date range: {facility_data['day'].min()} to {facility_data['day'].max()}")

# Prepare Prophet format
prophet_df = facility_data[['day', 'staffing_pressure', 'weighted_demand']].copy()
prophet_df.columns = ['ds', 'y', 'weighted_demand']
prophet_df = prophet_df.sort_values('ds')

print(f"\nProphet dataframe shape: {prophet_df.shape}")
print("First few rows:")
print(prophet_df.head())
```

```python
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    changepoint_prior_scale=0.05
)


model.add_regressor('weighted_demand')

model.fit(prophet_df)

print(" Prophet model trained successfully")
print(f"Regressors: {model.extra_regressors}")
```

```python
# Create future dataframe for next 7 days
future = model.make_future_dataframe(periods=7, freq='D')
print(f"Future dataframe shape: {future.shape}")
print(f"Last date: {future['ds'].iloc[-1]}")

# Need to provide future values for weighted_demand regressor
# Use average of last 7 days as simple forecast
last_7_days_demand = facility_data['weighted_demand'].iloc[-7:].values
future_demand = np.mean(last_7_days_demand)
print(f"Using weighted_demand forecast: {future_demand:.2f} (avg of last 7 days)")

# Add regressor values to future dataframe
future['weighted_demand'] = future_demand

# Make forecast
forecast = model.predict(future)
print(" Forecast generated")

# Show forecast columns
print(f"\nForecast columns: {forecast.columns.tolist()}")
print(f"\nLast 7 days forecast:")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(7))
```

```python
# Calculate threshold as 75th percentile of historical staffing_pressure
pressure_threshold = understaffed_df['staffing_pressure'].quantile(0.75)
print(f"Understaffing threshold (75th percentile): {pressure_threshold:.2f}")

# Check current facility's historical pressure vs threshold
facility_avg_pressure = facility_data['staffing_pressure'].mean()
print(f"Facility {first_facility} average pressure: {facility_avg_pressure:.2f}")
print(f"Currently understaffed? {facility_avg_pressure > pressure_threshold}")
```

```python
# Get the next 7 days forecast (last 7 rows)
next_7_days = forecast.tail(7).copy()

# Convert forecasted pressure to understaffing probability
# Using sigmoid function centered at threshold
def pressure_to_probability(pressure, threshold):
    # Sigmoid: 1 / (1 + exp(-k*(pressure/threshold - 1)))
    # k controls steepness, higher = sharper transition
    k = 4  # Adjustable parameter
    x = pressure / threshold
    return 1 / (1 + np.exp(-k * (x - 1)))

# Calculate probabilities
next_7_days['understaffing_probability'] = next_7_days['yhat'].apply(
    lambda x: pressure_to_probability(x, pressure_threshold)
)

# Mark as predicted understaffed if probability > 50%
next_7_days['predicted_understaffed'] = next_7_days['understaffing_probability'] > 0.5

print("Next 7 days forecast with understaffing probabilities:")
print(next_7_days[['ds', 'yhat', 'understaffing_probability', 'predicted_understaffed']])
```

```python
def forecast_facility_staffing(facility_data, periods=7):
    """Forecast staffing pressure for a single facility"""
    
    # Prepare data
    prophet_df = facility_data[['day', 'staffing_pressure', 'weighted_demand']].copy()
    prophet_df.columns = ['ds', 'y', 'weighted_demand']
    prophet_df = prophet_df.sort_values('ds')
    
    # Train model
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        changepoint_prior_scale=0.05
    )
    model.add_regressor('weighted_demand')
    model.fit(prophet_df)
    
    # Create future
    future = model.make_future_dataframe(periods=periods, freq='D')
    
    # Forecast weighted_demand (simple: average of last 7 days)
    last_demand = facility_data['weighted_demand'].iloc[-7:].mean()
    future['weighted_demand'] = last_demand
    
    # Predict
    forecast = model.predict(future)
    
    return forecast.tail(periods)

print(" Forecast function created")
```

```python
# Dictionary to store all forecasts
all_forecasts = {}

# Forecast for each facility
for facility_id in facilities:
    facility_data = understaffed_df[understaffed_df['facility_id'] == facility_id].copy()
    
    if len(facility_data) >= 30:  # Minimum data requirement
        try:
            forecast = forecast_facility_staffing(facility_data, periods=7)
            forecast['facility_id'] = facility_id
            
            # Calculate probabilities
            forecast['understaffing_probability'] = forecast['yhat'].apply(
                lambda x: pressure_to_probability(x, pressure_threshold)
            )
            forecast['predicted_understaffed'] = forecast['understaffing_probability'] > 0.5
            
            all_forecasts[facility_id] = forecast
            
            print(f" Forecast complete for {facility_id}")
            
        except Exception as e:
            print(f" Error forecasting {facility_id}: {str(e)}")
    else:
        print(f" Skipping {facility_id}: insufficient data ({len(facility_data)} records)")

print(f"\nTotal facilities forecasted: {len(all_forecasts)}")
```

```python
def predict_understaffing(facility_id, new_data_df, days_ahead=7):
    """
    Predict understaffing for a facility using new data
    
    Parameters:
    - facility_id: str, facility identifier
    - new_data_df: DataFrame with columns matching original schema
    - days_ahead: int, number of days to forecast (default: 7)
    
    Returns:
    - DataFrame with daily predictions
    """
    
    # Load the model
    model_path = f"saved_models/prophet_model_{facility_id}.pkl"
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"Model for {facility_id} not found at {model_path}")
    
    with open(model_path, 'rb') as f:
        model = pickle.load(f)
    
    # Calculate staffing pressure
    new_data_df['staffing_pressure'] = (
        new_data_df['staff_to_patient_ratio'] * 
        new_data_df['weighted_demand']
    )
    
    # Prepare for Prophet
    prophet_df = new_data_df[['day', 'staffing_pressure', 'weighted_demand']].copy()
    prophet_df.columns = ['ds', 'y', 'weighted_demand']
    prophet_df = prophet_df.sort_values('ds')
    
    # Create future dataframe
    future = model.make_future_dataframe(periods=days_ahead, freq='D')
    
    # Add regressor values (use last 7-day average)
    last_demand = new_data_df['weighted_demand'].iloc[-7:].mean()
    future['weighted_demand'] = last_demand
    
    # Predict
    forecast = model.predict(future)
    
    # Load threshold
    with open('saved_models/deployment_config.json', 'r') as f:
        config = json.load(f)
    
    threshold = config['threshold']
    
    # Convert to probabilities
    future_forecast = forecast.tail(days_ahead).copy()
    future_forecast['understaffing_probability'] = future_forecast['yhat'].apply(
        lambda x: pressure_to_probability(x, threshold)
    )
    future_forecast['predicted_understaffed'] = (
        future_forecast['understaffing_probability'] > 0.5
    )
    
    return future_forecast[['ds', 'yhat', 'understaffing_probability', 'predicted_understaffed']]

print("\n Prediction function created: predict_understaffing()")
print("  Usage: predictions = predict_understaffing('HF_L4_004', new_data, days_ahead=7)")
```

### Final Step

```python
# Calculate daily risk metrics
daily_metrics = []

for facility_id in understaffed_df['facility_id'].unique():
    facility_data = understaffed_df[understaffed_df['facility_id'] == facility_id].copy()
    
    # Sort by date
    facility_data = facility_data.sort_values('day')
    
    # Calculate 7-day rolling averages
    facility_data['staff_ratio_7d_avg'] = facility_data['staff_to_patient_ratio'].rolling(7).mean()
    facility_data['demand_7d_avg'] = facility_data['weighted_demand'].rolling(7).mean()
    facility_data['occupancy_7d_avg'] = facility_data['beds_occupancy_ratio'].rolling(7).mean()
    
    # Calculate staffing pressure (your model's target)
    facility_data['staffing_pressure'] = (
        facility_data['staff_to_patient_ratio'] * 
        facility_data['weighted_demand']
    )
    
    # Determine risk level for each day
    threshold = 97.03  # From your model
    facility_data['risk_score'] = facility_data['staffing_pressure'] / threshold
    facility_data['risk_level'] = pd.cut(
        facility_data['risk_score'],
        bins=[0, 0.7, 0.9, 1.1, float('inf')],
        labels=['Low', 'Medium', 'High', 'Critical']
    )
    
    # Add to daily metrics
    for _, row in facility_data.iterrows():
        daily_metrics.append({
            'facility_id': facility_id,
            'date': row['day'],
            'staff_ratio': row['staff_to_patient_ratio'],
            'weighted_demand': row['weighted_demand'],
            'occupancy': row['beds_occupancy_ratio'],
            'staffing_pressure': row['staffing_pressure'],
            'risk_score': row['risk_score'],
            'risk_level': row['risk_level'],
            'understaffed_flag': row['staffing_pressure'] > threshold
        })

daily_df = pd.DataFrame(daily_metrics)
```

```python
# Calculate latest status for each facility
facility_summary = []

for facility_id in understaffed_df['facility_id'].unique():
    facility_data = understaffed_df[understaffed_df['facility_id'] == facility_id]
    
    # Latest day
    latest = facility_data.iloc[-1]
    
    # Last 7 days stats
    last_7 = facility_data.tail(7)
    
    facility_summary.append({
        'facility_id': facility_id,
        'last_date': latest['day'],
        
        # Current metrics
        'current_staff_ratio': latest['staff_to_patient_ratio'],
        'current_demand': latest['weighted_demand'],
        'current_occupancy': latest['beds_occupancy_ratio'],
        'current_pressure': latest['staff_to_patient_ratio'] * latest['weighted_demand'],
        
        # 7-day trends
        'trend_staff_ratio': last_7['staff_to_patient_ratio'].mean(),
        'trend_demand': last_7['weighted_demand'].mean(),
        'trend_occupancy': last_7['beds_occupancy_ratio'].mean(),
        
        # Risk metrics
        'risk_score': (latest['staff_to_patient_ratio'] * latest['weighted_demand']) / 97.03,
        'understaffed_days_last_7': (last_7['staff_to_patient_ratio'] * last_7['weighted_demand'] > 97.03).sum(),
        'avg_risk_last_7': (last_7['staff_to_patient_ratio'] * last_7['weighted_demand']).mean() / 97.03
    })

summary_df = pd.DataFrame(facility_summary)

# Add risk levels
summary_df['risk_level'] = pd.cut(
    summary_df['risk_score'],
    bins=[0, 0.7, 0.9, 1.1, float('inf')],
    labels=['Low', 'Medium', 'High', 'Critical']
)

# Add trend direction
summary_df['trend'] = np.where(
    summary_df['current_pressure'] > summary_df['trend_staff_ratio'] * summary_df['trend_demand'],
    ' Rising',
    np.where(
        summary_df['current_pressure'] < summary_df['trend_staff_ratio'] * summary_df['trend_demand'],
        ' Falling',
        ' Stable'
    )
)
```

```python
# Save files for Power BI
daily_df.to_csv('daily_metrics.csv', index=False)
summary_df.to_csv('facility_summary.csv', index=False)

# Create calendar table for time intelligence
date_range = pd.date_range(
    start=daily_df['date'].min(),
    end=daily_df['date'].max(),
    freq='D'
)
calendar_df = pd.DataFrame({'date': date_range})
calendar_df['year'] = calendar_df['date'].dt.year
calendar_df['quarter'] = calendar_df['date'].dt.quarter
calendar_df['month'] = calendar_df['date'].dt.month
calendar_df['week'] = calendar_df['date'].dt.isocalendar().week
calendar_df['day_of_week'] = calendar_df['date'].dt.day_name()
calendar_df['is_weekend'] = calendar_df['date'].dt.dayofweek.isin([5, 6])

calendar_df.to_csv('calendar.csv', index=False)

print("Power BI files created:")
print(f"1. daily_metrics.csv - {len(daily_df)} rows")
print(f"2. facility_summary.csv - {len(summary_df)} facilities")
print(f"3. calendar.csv - {len(calendar_df)} dates")
```
